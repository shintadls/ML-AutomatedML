```python
import pandas as pd
import numpy as np
import autosklearn
from sklearn.model_selection import train_test_split
import time
```

```python
# reading files
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

# **Preprocessing**

```python
train_data
```

```python
train_data.info()
```

Turn diagnosis value into numeric value

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
train_data['diagnosis'] = le.fit_transform(train_data['diagnosis'])
```

```python
train_data.info()
```

The object type value needs to be replaced by float value, but it couldn't because it has '?' value

Drop rows with '?' value

```python
train_data = train_data[train_data.concavity_mean != '?']
```

```python
train_data[['concavity_mean', 'concave points_mean', 'concavity_se', 'concave points_se', 'concavity_worst', 'concave points_worst']] = train_data[['concavity_mean', 'concave points_mean', 'concavity_se', 'concave points_se', 'concavity_worst', 'concave points_worst']].apply(pd.to_numeric)
```

```python
train_data.info()
```

```python
test_data
```

```python
test_data.info()
```

# **Split Data into x and y**

```python
X = train_data.loc[:, train_data.columns != 'diagnosis']
y = train_data.loc[:, train_data.columns == 'diagnosis']
```

```python
#### Use Test Train Split to divide into train and test
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

# **TPOT**

```python
!pip install tpot
```

```python
from tpot import TPOTClassifier
from tpot.config import classifier_config_dict
#from tpot import TPOTRegressor
```

  Model 1 - Default

```python
#tpot
start_time = time.time()

tpot1 = TPOTClassifier(generations=5, population_size=20, verbosity=2, random_state=42)
tpot1.fit(X_train, y_train)

print(f'Pipeline Accuracy: {tpot1.score(X_test, y_test) * 100:.2f}%')

print(tpot1.fitted_pipeline_)

print(f"\nRuntime: {time.time() - start_time:.2f} s")
```

Export Best Pipeline Model 1

```python
tpot1.export('AutoML_model1.py')
```
